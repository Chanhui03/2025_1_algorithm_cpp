```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import pandas as pd

!git clone https://github.com/taehojo/data.git

df = pd.read_csv("/content/data/pima-indians-diabetes3.csv")

threshold = 0.2
targetCol = 'diabetes'

corr_matrix = df.corr()[targetCol]
relevant_cols = corr_matrix[abs(corr_matrix) > threshold].index

df_filtered = df[relevant_cols]

filtered_x = df_filtered.iloc[:, 0:4]
filtered_y = df_filtered.iloc[:, 4]

filtered_model = Sequential()
filtered_model.add(Dense(12, input_dim=4, activation='relu'))
filtered_model.add(Dense(10, activation='relu'))
filtered_model.add(Dense(8, activation='relu'))
filtered_model.add(Dense(1, activation='sigmoid'))

filtered_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

filtered_model.fit(filtered_x, filtered_y, epochs=20, batch_size=16)
```
